```python
import matplotlib.pyplot as plt
import numpy as np
```

## heat map

```python
fig, axe = plt.subplots(figsize = (8, 6))

x_labels = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6']
y_labels = ['y1', 'y2', 'y3', 'y4', 'y5', 'y6']

axe.set_xticks(np.arange(len(x_labels)))
axe.set_yticks(np.arange(len(y_labels)))

axe.set_xticklabels(x_labels)
axe.set_yticklabels(y_labels)

values = np.array([ [0.5, 0.8, 0.9, 0.7, 0.6, 0.4],
                    [0.4, 0.7, 0.8, 0.6, 1.0, 0.3],
                    [0.3, 0.6, 0.7, 0.5, 0.4, 0.2],
                    [0.2, 0.5, 0.6, 0.4, 0.3, 0.1],
                    [1.0, 0.4, 0.9, 0.3, 0.2, 0.8],
                    [0.2, 0.3, 0.4, 0.2, 0.1, 0.6]])


im = axe.imshow(values)

# list 出所有的值
for i in range(len(y_labels)):
    for j in range(len(x_labels)):
        text = axe.text(j, i, values[i, j], ha='center', va='center', color='red')


axe.figure.colorbar(im, ax=axe) # 顯示 colorbar

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))

x = np.random.randint(10, size = 50)
y = np.random.randint(20, size = 50)
z = np.random.randint(30, size = 50)

axes = plt.axes(projection = '3d')
axes.scatter3D(x, y, z, color = 'red')

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))

x = np.linspace(-4, 4, 400)
y = np.linspace(-4, 4, 400)
x, y = np.meshgrid(x, y) # 產生(x,y)對數網格
z = np.cos(np.sqrt(x**2 + y**2))

axes = plt.axes(projection = '3d')
axes.plot_surface(x, y, z)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))

x = np.linspace(-4, 7, 400)
y = np.linspace(-1, 4, 400)
x, y = np.meshgrid(x, y) # 產生(x,y)對數網格
z = np.cos(np.sqrt(x**2 + y**2))

axes = plt.axes(projection = '3d')
axes.plot_surface(x, y, z)

surf = axes.plot_surface(x, y, z, cmap = plt.get_cmap('plasma'))
plt.colorbar(surf)
plt.show()
```
