```python
import matplotlib.pyplot as plt
import numpy as np
```

## 面積圖

```python
fig, axe = plt.subplots(figsize=(8, 5))

x = np.arange(-3, 5, 0.5)
y = x ** 3
axe.fill_between(x, y, color='blue', alpha=0.8)

plt.show()
```

```python
fig, axe = plt.subplots(figsize=(8, 5))

x = np.arange(-3, 5, 0.5)
y1 = x ** 3
y2 = x ** 4

# fill_between(x, y1, y2) 会填充 y1 和 y2 之间的区域
axe.fill_between(x, y1, y2, color='blue', alpha=0.8)

plt.show()
```

```python
fig, axe = plt.subplots(figsize=(8, 5))

x = np.arange(-5, 5, 0.005)
y = np.sin(x * np.pi)

# 顏色填充
axe.fill_between(x, y, where = (y > 0), color='blue', alpha=0.8)
axe.fill_between(x, y, where = (y < 0), color='green', alpha=0.8)

plt.show()
```

## 堆疊面積圖

```python
fig, axe = plt.subplots(figsize=(8, 5))

x = [1, 2, 3, 4, 5]
y = [1, 3, 5, 7, 9]
np.random.seed(60)

y1 = y + np.random.randint(15, 65, 1)
y2 = y + np.random.randint(1, 15, 5)
y3 = y + np.random.randint(6, 11, 5)
y4 = y + np.random.randint(1, 55, 5)
y5 = y + np.random.randint(1, 10, 5)
y6 = y + np.random.randint(3, 15, 1)

labels = ['A', 'B', 'C', 'D', 'E']
axe.stackplot(x, y1, y2, y3, y4, y5, y6, labels = ['1', '2', '3', '4', '5', '6'])

axe.set_xticks(x)
axe.set_xticklabels(labels)
axe.set_title('Stackplot Example')
axe.legend(loc='upper left')
plt.show()
```
