```python
import matplotlib.pyplot as plt
import numpy as np
```

## 折線圖

```python
x = np.linspace(-5, 5, 100)

fig, axe = plt.subplots()
axe.plot(x, np.sin(x), '--*b')
axe.plot(x + 0.5, np.sin(x), '--r')
axe.plot(x + 3, np.sin(x), x + 3.5, np.sin(x))
plt.show()
```

## scatter 散點圖

```python
fig, axe = plt.subplots(figsize=(12, 3))

rng = np.random.RandomState(60)
x = rng.randn(50)
y = rng.randn(50)

colors = rng.rand(50)
sizes = 1000 * rng.rand(50)

axe.grid()
axe.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap='viridis')
plt.show()
```
