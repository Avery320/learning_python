```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x_points = np.array([1, 8])
y_points = np.array([9, 20])

plt.plot(x_points, y_points)
plt.show()
```

```python
major_category = ['A', 'B', 'C', 'D', 'E']

job_openings = [12000, 20000, 33200, 41200, 6000]
degrees_awarded = [10000, 20000, 36000, 64000, 58000]

x_axis = np.arange(len(job_openings))
plt.figure(figsize=(10, 5))
plt.bar(x_axis - 0.2, job_openings, 0.4, label='Job')
plt.bar(x_axis + 0.2, degrees_awarded, 0.4, label='Grads')

plt.xticks(x_axis, major_category)
plt.ylabel('job openings')
plt.title('Job Openings vs Degrees Awarded')

plt.legend()
plt.show()
```

```python
import math

t = np.linspace(0, math.pi, 10000)
x = np.sin(t)
y = np.cos(t) + np.power(x, 2.0/3)

plt.scatter(x, y, c = y, cmap = plt.cm.Reds, edgecolors = 'none', s = 40)
plt.scatter(-x, y, c = y, cmap = plt.cm.Reds, edgecolors = 'none', s = 40)

plt.axis([-2, 2, -2, 2])
plt.title('go go final project', color = 'Red', fontsize = 36)
plt.show()
```
