```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x1 = np.linspace(-5, 5, 50) # 50 points between -5 and 5
y1 = 2 * x1
y2 = x1 ** 2 - 20

fig = plt.figure(figsize=(16, 4))

plot = fig.add_subplot(131)
plot.plot(x1, y1)
plot.plot(x1, y2, color='red', linewidth=2, linestyle='--')

# set the limits of the x and y axes
plot = fig.add_subplot(132)
plot.plot(x1, y1)
plot.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.xlim(-4, 4)
plt.ylim(-15, 5)

plot = fig.add_subplot(133)
plot.plot(x1, y1)
plot.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.xlim(-4, 2)
plt.ylim(-15, 0)

plt.show()
```

```python
fig, axe = plt.subplots()

axe.plot(x1, y1)
axe.plot(x1, y2, color='red', linewidth=2, linestyle='--')

plt.xlim(-4, 2)
plt.ylim(-15, 0)

plt.show()
```

```python
plt.plot(x1, y1)
plt.plot(x1, y2, color='red', linewidth=2, linestyle='--')

plt.xticks([-2, 0, 2, 4], ['-x', '0', 'x', '2x']) #, 
plt.tick_params( direction = 'out', length = 3, width = 2, colors = 'blue') #


plt.show()
```

```python
fig, axe = plt.subplots()

y3 = np.tanh(x1) + 0.5
y4 = np.sin(x1) - 0.2

axe.plot(x1, y1, color = (0.1, 0.2, 0.5), linewidth = 2)
axe.plot(x1, y2, color = "#6F4F3F", label = 'hex', linewidth = 2)
axe.plot(x1, y3, color = 'r', label = 'rgb', linewidth = 2)
axe.plot(x1, y4, color = 'green', label = 'color name', linewidth = 2)
axe.legend()
plt.show()
```
