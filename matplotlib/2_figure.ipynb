```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# 第一種方法

x1 = np.linspace(-5, 5, 50) # 50 points between -5 and 5
y1 = 2 * x1
y2 = x1 ** 2 - 20

plt.figure(figsize=(4, 4))
plt.plot(x1, y1)
plt.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.show()
# plt.savefig('plot.png') # save the plot as a png file
```

```python
# 第二種方法

fig, axe = plt.subplots()
axe.plot(x1, y1)
axe.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.show()
```

```python
plt.figure(figsize = (10, 4)) #整個圖的大小

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x1, y1)
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 4))

plot = fig.add_subplot(121)
plot.plot(x1, y1)
plot = fig.add_subplot(122)
plot.plot(x1, y2, color='red', linewidth=2, linestyle='--')
plt.show()
```
